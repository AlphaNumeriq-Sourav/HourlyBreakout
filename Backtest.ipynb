```python


def Strategy(df , BasketName , dateIndex):

    DailyData = df.resample('D' ).agg({
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last',
        'volume': 'sum',
    }).dropna()

    EndTime = df.index[0] - timedelta(days=1)
    AvergaHoldingBars = 4
    for dates in dateIndex:
        
        if dates < EndTime :
            continue
        AverageVolume21days = DailyData[:dates].iloc[-21:].volume.mean()
        LastOneHourReturn = ((df[:dates].iloc[-2]['close'] - df[:dates].iloc[-2]['open'] )/df[:dates].iloc[-2]['open']) * 100
        
        if (AverageVolume21days >1000000) and (9 > LastOneHourReturn > 1):
            EntryPrice = df.loc[dates,'open']
            EntryTime = df.loc[dates:].index[0]
            ExitPrice = df.loc[dates:].iloc[AvergaHoldingBars]['close']
            EndTime = df.loc[dates:].iloc[:AvergaHoldingBars].index[-1]
            DrawDown = df.loc[dates:].iloc[:AvergaHoldingBars].low.min() - EntryPrice
            ExitType = 'SL'
            
            if ExitPrice > EntryPrice:
                ExitType = 'TP'
            
            Orders.loc[len(Orders)] = [1 , 'Long' , EntryTime , EndTime , EntryPrice , ExitPrice , ExitType , 0,0,DrawDown , BasketName , LastOneHourReturn]
            print(f'Entry at {dates} for {BasketName}')
            
            
Count = 0       
for file in files:
    sdf = pd.read_csv(file)
    sdf.date = pd.to_datetime(sdf.date)
    sdf.set_index('date' , inplace=True)
    dateIndex = sdf[(sdf.index > starting_period) & ((sdf.index < ending_period) )].index
    if sdf.index[0] > datetime(2019,1,1):
        dateIndex = dateIndex[30:]
    #try:
    Strategy(sdf, Symbols[Count],dateIndex)
    # except Exception as e:
    #     print(f'Error at {Symbols[Count]} : {e}')
    Count+=1
    print(f'\n\n Done Iteration : {Count}')
```
